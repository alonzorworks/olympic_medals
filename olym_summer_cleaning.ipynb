```python
# Cleaning the olympic data 
import pandas as pd 
import numpy as np
```

```python
# Notice the split creates two seperate columns. 
# The all cap name is the 
df["Athlete"].str.split(",", n = 1, expand = True )
```

```python
# Make the Null values readable by python
# Let;s take a look at the df.
df.replace(to_replace = "nan", value = np.nan)
```

```python
# Now we can save over the old null value.
df.replace(to_replace = "nan", value = np.nan, inplace = True)
```

```python
# Dropping nulls can allow conversion of floats to ints without error. 
# However it is better to use nulls that are compatible with integers when ever possible. 
df.dropna(inplace = True)
```

```python
# Converts Year to an integer. 
df.Year = df.Year.astype(int)
```

```python
df.head()
```

```python
medals = df.Medal.unique()
medals
```

```python
# Conditional statement to make the number of each medal countable based on the strings. 
# Use Numpy to achieve this.
# Archetype code.
# Notice how the first entry of Medals is silver.
# This was placed with an integer 1
np.where(df.Medal == "Silver", 1 ,df.Medal)
```

```python
# Making a Number for Each Medal Pythonically 
# The code iterates over the types of medal stored in the metal list derrived from the df.Medal.unique()
# From these values we can have the code mark a 1 in a column named after the actual medal.
# If there is no match we can simply just mark the non-

for i in medals:
    #df[f"{i}"] = 0
    df[f"{i}"] = np.where(df.Medal == f"{i}", 1, 0)
    
```
